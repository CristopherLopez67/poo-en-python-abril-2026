class Animal:
   lenguaje python crear la clase "animal" con los atributos "nombre", "raza" crear el constructor animal, y las funciones ladrar(), correr(), crear los objetos pitbull, chihuahua. nivel principiante, no es interactivo, proporciona tu los datos

```python
class Animal:
    # Constructor: inicializa los atributos al crear el objeto
    def __init__(self, nombre, raza):
        self.nombre = nombre
        self.raza = raza

    # Función para ladrar
    def ladrar(self):
        print(f"{self.nombre} (de raza {self.raza}) dice: ¡Guau, guau!")

    # Función para correr
    def correr(self):
        print(f"{self.nombre} está corriendo rápidamente. 👽")
    def morder(self):
        print(f"{self.nombre} está mordiendo a la vecina 👺")

# Creación de los objetos (instancias de la clase Animal)
pitbull = Animal("Rocco", "Pitbull")
chihuahua = Animal("Taco", "Chihuahua")
# creamos al objeto perro calejero con datos
perrocallejero = Animal("solovino", "chafa")
# Uso de los objetos
pitbull.ladrar()
pitbull.correr()

print("-" * 20)

chihuahua.ladrar()
chihuahua.correr()

print("-" * 20)

perrocallejero.ladrar()
perrocallejero.correr()
perrocallejero.morder()
print("programa realizado por Cristopher Lopez NL = 33")
```
